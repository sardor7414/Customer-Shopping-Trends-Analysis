```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Fayl archive papkasi ichida bo'lgani uchun yo'lni shunday yozamiz
file_path = 'archive/shopping_trends.csv'

try:
    df = pd.read_csv(file_path)
    print("Muvaffaqiyatli yuklandi!")
    print(f"Jami xaridlar soni: {df.shape[0]}")
    display(df.head())
except FileNotFoundError:
    print("Xato: Fayl hali ham topilmadi. Papka nomini tekshiring.")
```

```python
plt.figure(figsize=(15, 6))

# 1. Jins bo'yicha taqsimot
plt.subplot(1, 2, 1)
sns.countplot(x='Gender', data=df, palette='viridis')
plt.title('Mijozlarning jinsi bo\'yicha soni')

# 2. Yosh bo'yicha taqsimot
plt.subplot(1, 2, 2)
sns.histplot(df['Age'], bins=20, kde=True, color='teal')
plt.title('Mijozlarning yoshi bo\'yicha taqsimoti')

plt.show()
```

```python
plt.figure(figsize=(14, 7))

# 1. Kategoriyalar bo'yicha umumiy daromad
plt.subplot(1, 2, 1)
category_revenue = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
category_revenue.plot(kind='bar', color='orange')
plt.title('Kategoriyalar bo\'yicha daromad (USD)')
plt.ylabel('Jami summa')

# 2. Mavsumlar bo'yicha xaridlar
plt.subplot(1, 2, 2)
sns.countplot(x='Season', data=df, hue='Gender', palette='coolwarm')
plt.title('Mavsumlar va Jins bo\'yicha xaridlar')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Obuna holati bo'yicha o'rtacha xarid summasini tekshiramiz
sns.barplot(x='Subscription Status', y='Purchase Amount (USD)', data=df, palette='Set2')
plt.title('Obuna holati bo\'yicha o\'rtacha xarid summasi')

plt.show()

# Aniq raqamlarda ko'rish
print(df.groupby('Subscription Status')['Purchase Amount (USD)'].mean())
```
